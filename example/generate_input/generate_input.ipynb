```python
import pandas as pd
import numpy as np
from pymatgen.core.structure import Structure, Lattice
```

### Load Structure from POSCAR of CIF

#### Load POSCAR

```python
# Read structure from POSCAR file
poscar_path = "./POSCAR"  # Replace with your POSCAR file path
structure = Structure.from_file(poscar_path)
```

#### Load CIF

```python
# Read structure from CIF file
cif_path = "structure.cif"  # Replace with your CIF file path
structure = Structure.from_file(cif_path)
```

#### Create pymatgen Structure from cell, positions, and numbers

```python
cell = [[0.0, 3.1212, 3.1212], [3.1212, 0.0, 3.1212], [3.1212, 3.1212, 0.0]]
positions = [[0.75, 0.75, 0.75], [0.5, 0.5, 0.5], [0.25, 0.25, 0.25], [0.0, 0.0, 0.0]]
numbers = [22, 22, 25, 13]

lattice = Lattice(cell)
structure = Structure(lattice, numbers, positions)
```

#### Display Structure information

```python
# Display structure information
print(f"Formula: {structure.composition}")
print(f"Number of atoms: {len(structure)}")
print(f"Lattice parameters:")
print(f"  a={structure.lattice.a:.4f}, b={structure.lattice.b:.4f}, c={structure.lattice.c:.4f}")
print(f"  α={structure.lattice.alpha:.2f}°, β={structure.lattice.beta:.2f}°, γ={structure.lattice.gamma:.2f}°")
print(f"\nStructure:\n{structure}")
```

### Generate new structures by replacing atoms with other elements

```python
elements0 = ["V", "Cr", "Mn", "Fe", "Co", "Ni", "Cu", "Zn"]
elements1 = ["V", "Cr", "Mn", "Fe", "Co", "Ni", "Cu", "Zn"]
elements2 = ["V", "Cr", "Mn", "Fe", "Co", "Ni", "Cu", "Zn"]
elements3 = ["Ga", "Ge", "As", ]
```

```python
# Generate all combinations of elements and replace atoms in structure
from itertools import product

elements_lists = [elements0, elements1, elements2, elements3]

# Generate all combinations
structures_list = []    #where new structures are stored
combination_count = 0

for combination in product(*elements_lists):
    # Create a copy of the original structure
    modified_structure = structure.copy()
    
    # Replace atoms at each position with elements from the combination
    for pos, element in enumerate(combination):
        modified_structure[pos] = element
    
    structures_list.append(modified_structure)
    combination_count += 1
    
    # print progress info
    if combination_count <= 5 or combination_count % 100 == 0:
        print(f"Created structure {combination_count}: {combination}")

print(f"\nTotal structures created: {len(structures_list)}")
```

### Save structures to csv file

```python
# Store structures to a pandas dataframe
data_list = []

for idx, modified_structure in enumerate(structures_list):
    cell      = str(modified_structure.lattice.matrix.tolist())
    positions = str(modified_structure.frac_coords.tolist())
    numbers   = str(list(modified_structure.atomic_numbers))
    composition    = str(modified_structure.composition.hill_formula.replace(" ", ""))
    
    data_list.append({
        'index': idx,
        'composition': composition,
        'cell':      cell,
        'positions': positions,
        'numbers':   numbers
    })

# Create dataframe
df_structures = pd.DataFrame(data_list)
print(f"Dataframe shape: {df_structures.shape}")
print(f"\nFirst few rows:")
print(df_structures.head())

# Save dataframe to CSV
output_csv_path = "generated_structures.csv"
df_structures.to_csv(output_csv_path, index=False)
```
